find top 10 popular songs in this dataset

```python
top_ten = sp_tracks[sp_tracks['popularity']>=90]  # filter
top_ten.sort_values('popularity',ascending=False,inplace = True)  # sort
top_ten[['name','popularity','artists']][1:10]
```

Find top 10 long duration songs in this dataset

```python
sp_tracks.head()
```

```python
top_ten_duration = sp_tracks[sp_tracks['duration_ms']>=1000*60*10]
top_ten_duration.sort_values('duration_ms',ascending=False,inplace=True)
top_ten_duration[['name','duration_ms','artists']][1:10]
```
